# Carregar as bibliotecas

```python
import geopandas as gpd
import pandas as pd
```

# Ler os arquivos

```python
gdf = gpd.read_file('./data/base/PARBR2018_BATER.shp')

table = pd.read_excel(io = './data/VariaveisMorador/PARBR2018_VariaveisMorador.xls', 
                      sheet_name="SQL Results")

dicionario_variaveis = pd.read_excel(io = './data/VariaveisMorador/PARBR2018_VariaveisMorador.xls', 
                           sheet_name="Dicion√°rioVariaveis")
```

```python
gdf['GEO_BATER'] = gdf.GEO_BATER.astype(str)
table['COD_BATER'] = table.COD_BATER.astype(str)
```

```python
gdf = pd.merge(
    gdf,
    table,
    how="inner",
    right_on="COD_BATER",
    left_on="GEO_BATER"
)
```
